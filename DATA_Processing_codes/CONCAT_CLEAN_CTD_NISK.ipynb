```python
#PULL UP ALL THE EXCELL FILES 
files = glob.glob('../../../Data_Thesis/CTD/ROS_SUMMS/*GOM21FA*.csv')
#files = glob.glob('../../../OOC_Conferace_data/Fe/*FeT*.xlsx')
#files = glob.glob('../../../IRL22SU/TOC/*.xlsx')
files

#make the path where you want to save eveything 
path_save = '../../../Data_Thesis/CTD/ROS_SUMMS/GOM'
files
```

```python
#make function 
def GET_DATA(name):
    df = pd.read_csv(name, sep=',', index_col=0)
    df = pd.DataFrame(df)
    return df
```

```python
#add it all up 
frames = []
for f in files:
    df = GET_DATA(f) # runs the processing function 
    ID = os.path.basename(f)[:-4].split("_")
    camp = ID[1:2]
    station = ID[3:4]
    samp_ID = ID[2:3]
    df['camp'] = camp[0]
    df['station'] = station[0]
    df['samp_ID'] = samp_ID[0]
    frames.append(df) # adds the new col from each scan to a part of frames 
dffa = pd.concat(frames, axis=0, ignore_index=False)
dffa
```
