```python
#make function 
def GET_DATA(name):
    df = pd.read_csv(name, sep=',', index_col=0)
    df = pd.DataFrame(df)
    return df
```

```python
#add it all up 
frames = []
for f in files:
    df = GET_DATA(f) # runs the processing function  
    frames.append(df) # adds the new col from each scan to a part of frames 
df = pd.concat(frames, axis=0, ignore_index=False)
df
```

```python
#clean up messy collumns drop the extra 
df["Notes"] = df["Notes"].fillna(df["Unnamed: 10"])
df = df.drop(['Unnamed: 10'], axis=1)

#add the 2 depth cols together drop the extra 
df["Depth"] = df["Depth"].fillna(df["Depth (m)"])
df = df.drop(["Depth (m)"], axis=1)

#put it in a nicer order 
df = df[['Treatment','Campaign','Date','Location','Depth','Time','NPOC Conc (mg/L)','TN Conc (mg/L)','NPOC Conc (uM)','TN Conc (uM)','Notes']]
df
```

```python
#save it for later 
df.to_csv(path_save)
```
