```python
import pandas as pd #the csv reading library
import matplotlib.pyplot as plt #the ploting library
import glob #to get file names
import numpy as np
from pandas import Series
import seaborn as sns
```

```python
files = glob.glob('../../../FLBAY_CDOM/*out*.csv')
#files = glob.glob('../../../Data_Thesis/CDOM/OUTPUTS/*GOM*.csv')
files
path_save = '../../../FLBAY_CDOM/outputs.csv'
#path_save = '../../../Data_Thesis/CDOM/CDOM_GOM_SPECTRA.csv'
files
```

```python
def GET_DATA(name):
    df = pd.read_csv(name, sep=',', index_col=0)
    df = pd.DataFrame(df)
    return df
```

```python
frames = []
for f in files:
    df = GET_DATA(f) # runs the processing function  
    frames.append(df) # adds the new col from each scan to a part of frames 
df = pd.concat(frames, axis=0, ignore_index=False)
df
```

```python
#if it is a blank make it a nan 
df = (df.replace(r'^\s*$', np.nan, regex=True))       
df
```

```python
#make a col to play with 
df['sample_ID'] = df.index
df['sample_ID'] = df['sample_ID'].astype(str)

#clean up odd stuff left over from proccessing 
df_six = df.loc[df['sample_ID'].str.len()==6]
df_six.index = df_six['sample_ID'] 

df_eight = df.loc[df['sample_ID'].str.len()==8]
df_eight['sample_ID'] = df_eight['sample_ID'].str[2:]
df_eight.index = df_eight['sample_ID'] 

#adds them together 
frames = [df_eight, df_six]
df = pd.concat(frames)

#take it out if it is not a number then delet the row and rename the index 
df = df[pd.to_numeric(df['sample_ID'], errors='coerce').notnull()]
df = df.rename(columns={"sample_ID": "sample"})
df.index = df['sample'] 
df = df.drop(['sample'], axis=1)

df
```

```python
#save it for later 
df.to_csv(path_save)
```
