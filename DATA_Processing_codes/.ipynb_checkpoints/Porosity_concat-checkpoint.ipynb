```python
import pandas as pd
import os
import glob

files = glob.glob('../../../Data_Thesis/Porosity/G*')  # Replace with the actual path of the folder containing CSV files

def process_csv(csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)
    
    # Check if the required columns exist in the DataFrame
    required_columns = ['porosity', 'Average Depth (cm)']
    missing_columns = [col for col in required_columns if col not in df.columns]
    
    if missing_columns:
        print(f"Skipping file '{csv_file}' due to missing columns: {missing_columns}")
        return None
    
    # Rename columns if they exist
    df = df.rename(columns={df.columns[df.columns.get_loc('porosity')]: 'porosity',
                            df.columns[df.columns.get_loc("Average Depth (cm)")]: 'depth'})
    
    # Extract the file name and split into 'camp' and 'Stn'
    file_name = os.path.splitext(os.path.basename(csv_file))[0]
    camp, stn = file_name.split('_', 1)
    
    # Add 'camp' and 'Stn' columns
    df['camp'] = camp
    df['Stn'] = stn
    
    # Select only the required columns
    df = df[['porosity', 'depth', 'camp', 'Stn']]
    
    return df

def process_all_csv_files(csv_files):    
    # Initialize an empty list to store DataFrames from all CSV files
    all_dfs = []
    
    # Process each CSV file and append the DataFrame to the list
    for csv_file in csv_files:
        df = process_csv(csv_file)
        if df is not None:
            all_dfs.append(df)
    
    # Concatenate all DataFrames into a single DataFrame
    combined_df = pd.concat(all_dfs, ignore_index=True)
    
    return combined_df

if __name__ == "__main__":
    result_df = process_all_csv_files(files)

result_df
```
