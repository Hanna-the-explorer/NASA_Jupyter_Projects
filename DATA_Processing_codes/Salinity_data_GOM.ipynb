```python
from math import sin, cos, sqrt, atan2, radians

# Function to calculate Haversine distance
def haversine_distance(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)

    # Radius of the Earth in kilometers
    earth_radius = 6371.0

    # Haversine formula
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad
    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = earth_radius * c

    return distance

# Assuming you have a DataFrame called 'df' with columns "Lat" and "Lon"
# and you want to calculate distances from the point (29.154152, -89.239646)

# Latitude and Longitude of the reference point
ref_lat = 29.154152
ref_lon = -89.239646

# Calculate the distances using apply and lambda function
df['Distance_KM'] = df.apply(lambda row: haversine_distance(ref_lat, ref_lon, row['Lat'], row['Lon']), axis=1)

df
```

```python
df_locs = df_wc.groupby('code')['Depth','Site','Project'].max().reset_index()

#get WC bottom data 
grouped_data = df_wc.groupby(['code'])
sal_values = {}

for code, group in grouped_data:
    highest_row = group.loc[group['Depth'] == group['Depth'].max()]
    sal_values[code] = highest_row['Salinity'].iloc[0]

#get pw mean data 
grouped_data = df_pw.groupby(['code'])
mean_pw_values = {}

for code, group in grouped_data:
    mean_pw_values[code] = group['Salinity'].mean()
    
#get Lat mean data 
grouped_data = df_pw.groupby(['code'])
Lat_values = {}

for code, group in grouped_data:
    Lat_values[code] = group['Lat'].mean()
    
#get Lon mean data 
grouped_data = df_pw.groupby(['code'])
Lon_values = {}

for code, group in grouped_data:
    Lon_values[code] = group['Lat'].mean()
    
#get Lon mean data 
grouped_data = df_pw.groupby(['code'])
KM_values = {}

for code, group in grouped_data:
    KM_values[code] = group['Distance_KM'].mean()

# create a new column with "NA" values for all rows
df_locs['sal'] = np.nan
df_locs['pw_mean_sal'] = np.nan
df_locs['Lat'] = np.nan
df_locs['Lon'] = np.nan
df_locs['KM'] = np.nan


# create a boolean mask to select only the rows where the Station is in sal00_values
mask = df_locs['code'].isin(sal_values.keys())
mask = df_locs['code'].isin(mean_pw_values.keys())
mask = df_locs['code'].isin(Lat_values.keys())
mask = df_locs['code'].isin(Lon_values.keys())
mask = df_locs['code'].isin(KM_values.keys())

# use the mask to select only those rows and add the sal00 values to the new column
df_locs.loc[mask, 'sal'] = [sal_values[code] for code in df_locs.loc[mask, 'code']]
df_locs.loc[mask, 'pw_mean_sal'] = [mean_pw_values[code] for code in df_locs.loc[mask, 'code']]
df_locs.loc[mask, 'Lat'] = [Lat_values[code] for code in df_locs.loc[mask, 'code']]
df_locs.loc[mask, 'Lon'] = [Lon_values[code] for code in df_locs.loc[mask, 'code']]
df_locs.loc[mask, 'KM'] = [KM_values[code] for code in df_locs.loc[mask, 'code']]


df_locs = df_locs.sort_values('Depth')

#df_locs
col_order = df_locs["code"].tolist()
col_maxd = df_locs["Depth"].tolist()
col_maxd = [round(x, 1) for x in col_maxd]
col_sal = df_locs["sal"].tolist()
col_sal = [round(x, 1) for x in col_sal]
col_site = df_locs["Site"].tolist()
col_camp = df_locs["Project"].tolist()


sub_num = len(df_locs)-1
width_num = len(df_locs)

df_locs
```

```python
import pandas as pd
import numpy as np
from scipy.interpolate import pchip_interpolate

# Assuming you have the DataFrame df_por and df_pw as described in the previous code snippet
# Assuming my_list, unique_codes2, and interp_porosity are defined as shown in the previous code snippet

# Initialize an empty DataFrame to store the integrated data
integrated_df = pd.DataFrame(columns=['code', 'Depth', 'porosity'])

for x in my_list:
    # Get the corresponding group from df_por based on 'code'
    porosity_group = df_por[df_por['code'] == unique_codes2[x]]
    pw_group = df_pw[df_pw['code'] == unique_codes2[x]]
    
    # Drop rows in pw_group where 'Depth' is less than 0
    pw_group = pw_group.drop(pw_group[pw_group['Depth'] < 0].index)
    
    X_por = np.asarray(porosity_group['porosity'], dtype=float).ravel()
    Y_dep = np.asarray(porosity_group['Depth'], dtype=float).ravel()
    Z_dep = np.asarray(pw_group['Depth'], dtype=float).ravel()
    
    # Create an interpolation function based on the 'Depth' and 'porosity' columns
    interp_porosity = pchip_interpolate(Y_dep, X_por, Z_dep)
    
    # Create a DataFrame to store the interpolated data
    interpolated_data = pd.DataFrame({
        'code': unique_codes2[x],
        'Depth': Z_dep,
        'porosity': interp_porosity
    })
    
    # Append the interpolated data to the integrated DataFrame
    integrated_df = pd.concat([integrated_df, interpolated_data], ignore_index=True)

# Now, integrated_df contains the interpolated data matched to the corresponding 'code' and 'Depth' values
df_pw = pd.merge(df_pw, integrated_df, on=['code', 'Depth'], how='left')
df_pw
```
